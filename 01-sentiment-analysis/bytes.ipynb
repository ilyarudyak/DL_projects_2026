# 01 Beazley

```python
# Create a bytes object from a string
a = b'hello'

# Check the type of `a`
print(type(a))  # Output: <class 'bytes'>

# Print the bytes object
print(a)  # Output: b'hello'

# Print individual bytes
print(list(a))  # Output: [104, 101, 108, 108, 111]
print([byte for byte in a])  # Output: [104, 101, 108, 108, 111]
print([chr(byte) for byte in a])  # Output: ['h', 'e', 'l', 'l', 'o']
print([hex(byte) for byte in a])  # Output: ['0x68', '0x65', '0x6c', '0x6c', '0x6f']
```

```python
# Specify bytes from a list of integers
b = bytes([0x68, 0x65, 0x6c, 0x6c, 0x6f])

# Print the bytes object
print(b)  # Output: b'hello'

# Print individual bytes
print(list(b))  # Output: [104, 101, 108, 108, 111]
print([byte for byte in b])  # Output: [104, 101, 108, 108, 111]
print([chr(byte) for byte in b])  # Output: ['h', 'e', 'l', 'l', 'o']
print([hex(byte) for byte in b])  # Output: ['0x68', '0x65', '0x6c', '0x6c', '0x6f']
```

```python
a = 'hello'
b = a.encode('utf-8')
print(b)  # Output: b'hello'
print([hex(byte) for byte in b])  # Output: ['0x68', '0x65', '0x6c', '0x6c', '0x6f']

c = b.decode('utf-8')
print(c)  # Output: hello
```

## 01 Unicode Foundations

```python
a_encoded = 'a'.encode('ASCII')
print(type(a_encoded))  # Output: <class 'bytes'>
print(len(a_encoded))  # Output: 1
```

```python
a_encoded_utf = 'a'.encode('UTF-8')
print(type(a_encoded_utf))  # Output: <class 'bytes'>
print(len(a_encoded_utf))  # Output: 1
print([hex(byte) for byte in a_encoded_utf])  # Output: ['0x61']

b = "Ä"
b_encoded_utf = b.encode('UTF-8')
print(type(b_encoded_utf))  # Output: <class 'bytes'>
print(len(b_encoded_utf))  # Output: 2
print([hex(byte) for byte in b_encoded_utf])  # Output: ['0xc3', '0x84']

c = "😊"
c_encoded_utf = c.encode('UTF-8')
print(type(c_encoded_utf))  # Output: <class 'bytes'>
print(len(c_encoded_utf))  # Output: 4
print([hex(byte) for byte in c_encoded_utf])  # Output: ['0xf0', '0x9f', '0x98', '0x8a']
```

## 02 Introducing Python String Tools

```python
# A bytes object really is a sequence of small integers, each of which is in the range 0…255
a = b'hello'
print(list(a))  # Output: [104, 101, 108, 108, 111]
print([byte for byte in a])  # Output: [104, 101, 108, 108, 111]
print(a[0])  # Output: 104

# We may create a bytes object from a list of integers, each of which is in the range 0…255
a2 = bytes([104, 101, 108, 108, 111])
print()
print(list(a2))  # Output: [104, 101, 108, 108, 111]
print([byte for byte in a2])  # Output: [104, 101, 108, 108, 111]
print(a2[0])  # Output: 104
```

```python
# Let's look at indexing and slicing of bytes objects
# Indexing returns an integer, NOT a bytes object
a = b'hello'
print(type(a[0]))  # Output: <class 'int'>

# Slicing returns a bytes object
print(type(a[0:2]))  # Output: <class 'bytes'>

# list() returns a list of integers, NOT a list of bytes objects or characters
al = list(a)
print(type(al), type(al[0]))  # Output: <class 'list'>, <class 'int'>
```

```python
# When printing a bytes object, Python displays the ASCII characters for the bytes that correspond 
# to printable ASCII characters, and it displays hexadecimal escape sequences for the bytes that 
# do not correspond to printable ASCII characters or are outside the ASCII range (0-127). 
a = bytes([104, 101, 108, 128, 255])
print(a)  # Output: b'hel\x80\xff'
```

```python
# Let's explore what are these printable ASCII characters
a = bytes(range(128))
with open('ascii_bytes.txt', 'w') as f:
    for i, byte in enumerate(a):
        if 32 <= byte <= 126:
            f.write(f"{i}: {byte} -> '{chr(byte)}'\n")
        else:
            f.write(f"{i}: {byte} -> non-printable\n")
```

```python
# Mapping of non-printable C0 Control Characters (0-31 and 127)
CONTROL_CHARS = {
    0: "NUL (null)", 1: "SOH (start of heading)", 2: "STX (start of text)",
    3: "ETX (end of text)", 4: "EOT (end of transmission)", 5: "ENQ (enquiry)",
    6: "ACK (acknowledge)", 7: "BEL (bell)", 8: "BS (backspace)",
    9: "TAB (horizontal tab)", 10: "LF (NL line feed, new line)", 11: "VT (vertical tab)",
    12: "FF (NP form feed, new page)", 13: "CR (carriage return)", 14: "SO (shift out)",
    15: "SI (shift in)", 16: "DLE (data link escape)", 17: "DC1 (device control 1)",
    18: "DC2 (device control 2)", 19: "DC3 (device control 3)", 20: "DC4 (device control 4)",
    21: "NAK (negative acknowledge)", 22: "SYN (synchronous idle)", 23: "ETB (end of trans. block)",
    24: "CAN (cancel)", 25: "EM (end of medium)", 26: "SUB (substitute)",
    27: "ESC (escape)", 28: "FS (file separator)", 29: "GS (group separator)",
    30: "RS (record separator)", 31: "US (unit separator)", 127: "DEL (delete)"
}

a = bytes(range(128))
with open('ascii_bytes_detailed.txt', 'w') as f:
    for byte in a:
        if 32 <= byte <= 126:
            # Printable
            f.write(f"{byte:3}: '{chr(byte)}'\n")
        else:
            # Non-printable: Show the official abbreviation AND the Python hex escape
            name = CONTROL_CHARS.get(byte, "unknown")
            # repr() shows escape sequences like \x00 or \n
            hex_esc = repr(bytes([byte]))[2:-1] 
            f.write(f"{byte:3}: {hex_esc:4} -> [{name}]\n")
```
